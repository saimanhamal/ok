## 4.1 Exercise on Functions

### Task 1: Unit Converter

```python
def convert_length(value, direction):
    """
    Convert between meters and feet.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'm_to_ft' or 'ft_to_m'.

    Returns:
        float: Converted value.
    """
    if direction == 'm_to_ft':
        return value * 3.28084
    elif direction == 'ft_to_m':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'm_to_ft' or 'ft_to_m'.")


def convert_weight(value, direction):
    """
    Convert between kilograms and pounds.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'kg_to_lbs' or 'lbs_to_kg'.

    Returns:
        float: Converted value.
    """
    if direction == 'kg_to_lbs':
        return value * 2.20462
    elif direction == 'lbs_to_kg':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'kg_to_lbs' or 'lbs_to_kg'.")


def convert_volume(value, direction):
    """
    Convert between liters and gallons.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'l_to_gal' or 'gal_to_l'.

    Returns:
        float: Converted value.
    """
    if direction == 'l_to_gal':
        return value * 0.264172
    elif direction == 'gal_to_l':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'l_to_gal' or 'gal_to_l'.")


def unit_converter():
    """
    Interactive unit converter for length, weight, and volume.
    Handles invalid input with try-except blocks.
    """
    print("Unit Converter")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kg <-> lbs)")
    print("3. Volume (liters <-> gallons)")

    try:
        choice = input("Choose conversion type (1/2/3): ").strip()
        value = float(input("Enter the value to convert: "))

        if choice == '1':
            direction = input("Direction? (m_to_ft / ft_to_m): ").strip()
            result = convert_length(value, direction)
            print(f"Result: {value} -> {result:.4f}")
        elif choice == '2':
            direction = input("Direction? (kg_to_lbs / lbs_to_kg): ").strip()
            result = convert_weight(value, direction)
            print(f"Result: {value} -> {result:.4f}")
        elif choice == '3':
            direction = input("Direction? (l_to_gal / gal_to_l): ").strip()
            result = convert_volume(value, direction)
            print(f"Result: {value} -> {result:.4f}")
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
    except ValueError as e:
        print(f"Error: {e}")


# --- Test without interactive input ---
print("=== Length: 10 meters to feet ===")
print(f"{convert_length(10, 'm_to_ft'):.4f} ft")

print("\n=== Weight: 70 kg to lbs ===")
print(f"{convert_weight(70, 'kg_to_lbs'):.4f} lbs")

print("\n=== Volume: 5 liters to gallons ===")
print(f"{convert_volume(5, 'l_to_gal'):.4f} gal")
```

### Task 2: Mathematical Operations on a List

```python
def compute_sum(numbers):
    """
    Compute the sum of a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Sum of the list.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return sum(numbers)


def compute_average(numbers):
    """
    Compute the average of a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Average of the list.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Find the maximum value in a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Maximum value.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return max(numbers)


def compute_minimum(numbers):
    """
    Find the minimum value in a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Minimum value.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return min(numbers)


def math_operations():
    """
    Interactive math operations on user-provided list of numbers.
    """
    print("Math Operations")
    print("1. Sum  2. Average  3. Maximum  4. Minimum")
    try:
        choice = input("Choose operation (1/2/3/4): ").strip()
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.split()]

        ops = {'1': compute_sum, '2': compute_average,
               '3': compute_maximum, '4': compute_minimum}

        if choice in ops:
            result = ops[choice](numbers)
            print(f"Result: {result}")
        else:
            print("Invalid choice.")
    except ValueError as e:
        print(f"Error: {e}")


# --- Test without interactive input ---
nums = [4, 7, 2, 9, 1, 5]
print(f"Numbers: {nums}")
print(f"Sum:     {compute_sum(nums)}")
print(f"Average: {compute_average(nums):.2f}")
print(f"Maximum: {compute_maximum(nums)}")
print(f"Minimum: {compute_minimum(nums)}")
```

## 4.2 Exercise on List Manipulation

```python
def extract_every_other(lst):
    """Return every other element starting from the first."""
    return lst[::2]

def get_sublist(lst, start, end):
    """Return sublist from start to end (inclusive)."""
    return lst[start:end+1]

def reverse_list(lst):
    """Reverse a list using slicing."""
    return lst[::-1]

def remove_first_last(lst):
    """Remove the first and last elements."""
    return lst[1:-1]

def get_first_n(lst, n):
    """Return the first n elements."""
    return lst[:n]

def get_last_n(lst, n):
    """Return the last n elements."""
    return lst[-n:]

def reverse_skip(lst):
    """Return every second element starting from second-to-last, going backward."""
    return lst[-2::-2]


# --- Tests ---
sample = [1, 2, 3, 4, 5, 6]
print(f"Original:            {sample}")
print(f"Every other:         {extract_every_other(sample)}")
print(f"Sublist [2..4]:      {get_sublist(sample, 2, 4)}")
print(f"Reversed:            {reverse_list(sample)}")
print(f"Remove first/last:   {remove_first_last(sample)}")
print(f"First 3:             {get_first_n(sample, 3)}")
print(f"Last 2:              {get_last_n(sample, 2)}")
print(f"Reverse skip:        {reverse_skip(sample)}")
```

## 4.3 Exercise on Nested List

```python
def flatten(lst):
    """Flatten a one-level nested list."""
    return [item for sublist in lst for item in sublist]

def access_nested_element(lst, indices):
    """Access element at given nested indices."""
    result = lst
    for idx in indices:
        result = result[idx]
    return result

def sum_nested(lst):
    """Recursively sum all elements in a nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

def remove_element(lst, elem):
    """Remove all occurrences of elem from nested list."""
    return [[x for x in sublist if x != elem] for sublist in lst]

def find_max(lst):
    """Find max in a nested list (any depth)."""
    def _flatten(l):
        for item in l:
            if isinstance(item, list):
                yield from _flatten(item)
            else:
                yield item
    return max(_flatten(lst))

def count_occurrences(lst, elem):
    """Count occurrences of elem in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

def deep_flatten(lst):
    """Flatten a deeply nested list of any depth."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

def average_nested(lst):
    """Calculate average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)


# --- Tests ---
print("1. Flatten:", flatten([[1, 2], [3, 4], [5]]))
print("2. Access nested element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1, 2]))
print("3. Sum nested:", sum_nested([[1, 2], [3, [4, 5]], 6]))
print("4. Remove element:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
print("5. Find max:", find_max([[1, 2], [3, [4, 5]], 6]))
print("6. Count occurrences:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
print("7. Deep flatten:", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
print("8. Average nested:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

### Problem 4: Matrix Operations

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

print("A =\n", A)
print("B =\n", B)

# 1. Prove A * A_inv = I
A_inv = np.linalg.inv(A)
product_AI = np.round(A @ A_inv, 6)
print("\n1. A · A⁻¹ =\n", product_AI, "  ✓ Identity")

# 2. Prove AB != BA
AB = A @ B
BA = B @ A
print("\n2. AB =\n", AB)
print("   BA =\n", BA)
print("   AB == BA?", np.allclose(AB, BA), "  ✓ Not equal")

# 3. Prove (AB)^T = B^T * A^T
ABT = (A @ B).T
BTAT = B.T @ A.T
print("\n3. (AB)ᵀ =\n", ABT)
print("   BᵀAᵀ =\n", BTAT)
print("   Equal?", np.allclose(ABT, BTAT), "  ✓ Proved")
```

#### Solving System of Linear Equations

```python
import numpy as np

# System:
# 2x - 3y +  z = -1
#  x -  y + 2z = -3
# 3x +  y -  z =  9

A_sys = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]], dtype=float)

B_sys = np.array([-1, -3, 9], dtype=float)

# Method 1: Manual inverse method
A_inv = np.linalg.inv(A_sys)
X_manual = A_inv @ B_sys
print("Inverse Method Solution:")
print(f"  x = {X_manual[0]:.4f}, y = {X_manual[1]:.4f}, z = {X_manual[2]:.4f}")

# Method 2: np.linalg.inv (same but explicit)
X_inv = np.linalg.inv(A_sys) @ B_sys
print("\nnp.linalg.inv Solution:")
print(f"  x = {X_inv[0]:.4f}, y = {X_inv[1]:.4f}, z = {X_inv[2]:.4f}")

# Verify
print("\nVerification A @ X =", np.round(A_sys @ X_manual, 4), " == B =", B_sys)
```
